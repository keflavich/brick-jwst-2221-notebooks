```python
# one-time correction: add visit metadata
import glob
from astropy.io import fits
from astropy.table import Table

basepath = '/blue/adamginsburg/adamginsburg/jwst/brick/'
for fn in glob.glob(f"{basepath}F*/*visit*exp*fits"):
    visit = fn.split("visit")[1][:3]
    exp = fn.split("exp")[1][:5]
    if 'skymodel' in fn or fn.endswith('psf.fits') or 'daofind' in fn:
        continue

    hdr = fits.getheader(fn, ext=1)
    if 'VISIT' in hdr and 'RAOFFSET' in hdr:
        continue

    with fits.open(fn, mode='update') as fh:
        fh[1].header['visit'] = visit

        ffn = fh[1].header['FILENAME']
        hdr = fits.getheader(ffn, ext=('SCI', 1))

        fh[1].header['RAOFFSET'] = hdr['RAOFFSET']
        fh[1].header['DEOFFSET'] = hdr['DEOFFSET']

    
    print(fn, visit, exp, Table.read(fn).meta['VISIT'], hdr['RAOFFSET'])
```
